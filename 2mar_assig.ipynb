Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of tools and functionalities for creating various types of plots and charts.

Matplotlib is used for data visualization, allowing users to represent their data in a visual format. It is commonly used in data analysis, scientific research, and data exploration tasks. With Matplotlib, you can create line plots, scatter plots, bar plots, histograms, pie charts, and many other types of visualizations.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line Plot: A line plot displays data points connected by straight lines. It is useful for visualizing trends and patterns in continuous data over a specific range.

Scatter Plot: A scatter plot represents individual data points as markers in a coordinate system. It is useful for exploring relationships or correlations between two numeric variables.

Bar Plot: A bar plot represents data using rectangular bars with lengths proportional to the values they represent. It is often used to compare different categories or groups.

Histogram: A histogram visualizes the distribution of a single variable by dividing it into bins and displaying the frequency or count of values falling into each bin. It provides insights into the underlying data distribution.

Pie Chart: A pie chart represents data as slices of a circular pie, where each slice represents a category and its size represents the proportion or percentage of that category in the whole. It is useful for displaying proportions or compositions of a dataset.

These are just a few examples of the plots that can be created using the Pyplot module of Matplotlib. Matplotlib offers a wide range of customization options to tailor the plots according to specific requirements.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot that displays the relationship or correlation between two numeric variables. It uses Cartesian coordinates to represent each data point as a point in a two-dimensional space. Each point on the scatter plot corresponds to a unique combination of values from the two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.title("Scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:


```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4= np.array([200, 350, 250, 550, 450, 150])
```

he subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or charts in a single figure window, arranged in a grid-like structure.

By using subplot(), you can specify the number of rows and columns in the grid, as well as the index of the specific subplot you want to create. This allows for easy organization and comparison of multiple plots in a single figure.

```python
import matplotlib.pyplot as plt 
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('Line 1')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Line 2')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Line 3')

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot, also known as a bar chart or bar graph, is a visualization that represents categorical data using rectangular bars. The length of each bar corresponds to the value or quantity associated with a particular category, allowing for easy comparison between different categories.

Bar plots are commonly used to display and compare discrete or categorical data, such as comparing sales figures for different products, comparing performance metrics of different companies, or visualizing survey results based on different response options.

Bar plots are used for the following purposes:

Comparison: Bar plots allow for quick visual comparison of values between different categories. They make it easy to identify the highest or lowest values and identify any significant differences.

Categorization: Bar plots help in categorizing data based on different groups or categories. They provide a clear representation of the distribution of data across various categories.

Trend analysis: Bar plots can also be used to analyze trends or changes over time by representing different time periods or intervals on the x-axis and the corresponding values on the y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("company profit")
plt.show()

plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("company profit")

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box-and-whisker plot, is a statistical visualization that displays the summary of a set of data values. It provides a compact way to represent the distribution of a dataset, showing the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum values.

Box plots are used for the following purposes:

Summary of data distribution: Box plots provide a visual summary of the main characteristics of a dataset, including the central tendency, spread, and presence of outliers.

Comparison of distributions: Box plots allow for easy comparison of distributions between different groups or categories, providing insights into differences in medians, spreads, and overall shapes of the distributions.

Identification of outliers: Box plots help in identifying potential outliers in the dataset, which are values that significantly deviate from the rest of the data.

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)

data=[box1,box2]

plt.boxplot(data)
plt.xticks([1,2] , ['box1' , 'box2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
